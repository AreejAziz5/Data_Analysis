### Data Cleaning & pre-processing

```python
df.isna().sum()
```

```python
df.fillna(df['Weight_kg'].mean())
df.fillna(df['Screen_Size_cm'].mean())
```

### Exploratory Data Analysis

```python
df1 = df.sample(n = 40)
df1
```

```python
plt.barh('Manufacturer','Price',data = df, height =0.5)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.bar(df['Manufacturer'], df['Price'], label='Price', color='blue', alpha=0.7)
plt.bar(df['Manufacturer'], df['Weight_kg'], label='Weight_kg', color='green', alpha=0.7)
plt.bar(df['Manufacturer'], df['Storage_GB_SSD'], label='Storage_GB_SSD', color='red', alpha=0.7)
plt.xlabel('Manufacturer')
plt.ylabel('Price, Weight_kg, and Storage_GB_SSD')
plt.title('Price, Weight_kg, and Storage_GB_SSD')
plt.legend()
plt.show()
```

```python
sns.set(font_scale=.60)
sns.countplot(hue='CPU_core', x='Manufacturer',data = df)
```

```python
sns.barplot(x ='GPU',data = df, y='Price')
```

### Positive Correlation between price and CPU_core

```python
sns.set_style('whitegrid')
sns.lmplot( x = 'CPU_core',y ='Price', data = df)
```

#### top 10 expensive laptop

```python
top10 = df.nlargest(10,'Price')
plt.figure(figsize=(20,8))
plt.bar(top10['Manufacturer'], top10['Price'],color='gold',alpha=0.6)
plt.ylabel('Price', fontsize = 10)
plt.xlabel("Manufacturer", fontsize = 10)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.title("Top 10 Expensive Laptop Manufacturer", fontsize = 20)
```

#### top 10 Cheep laptop

```python
top10 = df.nsmallest(10,'Price')
plt.figure(figsize=(20,8))
plt.bar(top10['Manufacturer'], top10['Price'],color='brown')
plt.ylabel('Price',fontsize = 15)
plt.xlabel("Manufacturer", fontsize = 15)
plt.title("Top 10 Cheap Laptop Manufacturer", fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

plt.show()
```
